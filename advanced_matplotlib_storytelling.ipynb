```python
# Advanced Data Visualization & Storytelling Dashboard

## Proje Amacı

#Bu projenin amacı yalnızca grafik çizmek değil, veriyi **hikâye olarak anlatabilen bir dashboard** oluşturmaktır.

#Şirket yöneticileri genellikle onlarca rapor okumak istemez.
#Tek bir sayfada:

#- satış trendlerini
#- ürün performanslarını
#- bölgesel dağılımları
#- müşteri davranışlarını anlayabilecekleri bir görselleştirme isterler.

#Bu nedenle **çoklu grafik içeren tek bir dashboard** tasarlanmıştır.


## Kullanılan Matplotlib Konuları

#Bu projede yalnızca aşağıdaki konular kullanılmıştır:

#- Subplots (2x2 dashboard düzeni)
#- Görsel geliştirmeler (grid, annotation, text)
#- İleri seviye görsellik
#- Farklı grafik türleri
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# Aylar
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Aylık satış verisi
sales = np.random.randint(200, 800, 12)

sales_df = pd.DataFrame({
    "Month": months,
    "Sales": sales
})

# Ürün satışları
products = ["Product A", "Product B", "Product C", "Product D"]
product_sales = np.random.randint(300,1000,4)

product_df = pd.DataFrame({
    "Product": products,
    "Sales": product_sales
})

# Bölgesel performans
regions = ["Marmara","Ege","Akdeniz","İç Anadolu","Karadeniz"]
region_sales = np.random.randint(200,900,5)

region_df = pd.DataFrame({
    "Region": regions,
    "Sales": region_sales
})

# Müşteri harcama dağılımı
customer_spending = np.random.normal(200,50,200)
```

```python
fig, axes = plt.subplots(2,2, figsize=(14,10))

fig.suptitle(
    "Company Performance Dashboard",
    fontsize=18,
    fontweight="bold"
)
```

```python
ax1 = axes[0,0]

ax1.plot(
    sales_df["Month"],
    sales_df["Sales"],
    marker="o",
    linewidth=2
)

ax1.set_title("Monthly Sales Trend")
ax1.set_ylabel("Sales")

ax1.grid(True, linestyle="--", alpha=0.6)

# En yüksek satış anotasyonu
max_index = sales_df["Sales"].idxmax()

ax1.annotate(
    "Peak Sales",
    (sales_df["Month"][max_index], sales_df["Sales"][max_index]),
    xytext=(0,15),
    textcoords="offset points",
    arrowprops=dict(arrowstyle="->")
)
```

```python
ax2 = axes[0,1]

ax2.bar(
    product_df["Product"],
    product_df["Sales"]
)

ax2.set_title("Product Sales Comparison")
ax2.set_ylabel("Sales")

ax2.grid(axis="y", linestyle="--", alpha=0.6)
```

```python
ax3 = axes[1,0]

ax3.scatter(
    region_df["Region"],
    region_df["Sales"],
    s=120
)

ax3.set_title("Regional Performance")
ax3.set_ylabel("Sales")

ax3.grid(True, linestyle="--", alpha=0.6)
```

```python
ax4 = axes[1,1]

ax4.hist(
    customer_spending,
    bins=20
)

ax4.set_title("Customer Spending Distribution")
ax4.set_xlabel("Spending")
ax4.set_ylabel("Frequency")

ax4.grid(True, linestyle="--", alpha=0.6)
```

```python
plt.tight_layout()
plt.show()
```

```python
# Dashboard Hikâye Akışı

#Dashboard soldan sağa ve yukarıdan aşağıya okunacak şekilde tasarlanmıştır.

#1️⃣ **Monthly Sales Trend (Line Plot)**  
#Satışların yıl boyunca nasıl değiştiğini gösterir.

#2️⃣ **Product Sales Comparison (Bar Plot)**  
#Hangi ürünün gelirde lider olduğunu açıkça gösterir.

#3️⃣**Regional Performance (Scatter Plot)**  
#Hangi bölgelerin satışta güçlü olduğunu ortaya koyar.

#4️⃣ **Customer Spending Distribution (Histogram)**  
#Müşteri harcamalarının nasıl dağıldığını gösterir.

#Histogram kullanılmasının nedeni:
#müşteri harcamalarının **ortalama etrafında nasıl dağıldığını görmek**.

#Bu grafikler birlikte incelendiğinde şirketin satış performansı
#tek bir dashboard üzerinden anlaşılabilir.
```
